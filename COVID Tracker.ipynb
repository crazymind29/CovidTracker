# This Data analysis is performed based on 26/04/2021 at 20:00.

```python
import pandas as pd
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df.head()
```

```python
df.columns
```

```python
data = pd.DataFrame(df, columns = ['sno', 'state_name', 'active', 'positive', 'cured', 'death', 'new_active', 
                                       'new_positive', 'new_cured', 'new_death','state_code', ])
```

```python
data
```

```python
### remove last row as represented sum of all features and remove also sno 

data = data.iloc[:-1, 1:]
```

```python
data.head()
```

```python
data.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
total_states = np.arange(len(data['state_name']))
```

```python
total_states
```

```python
data['active']
```

```python
max(data['positive'])
```

### Total Positive Cases Based on States in India

```python
from matplotlib.pyplot import figure

figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, data['positive'], align='center', alpha=0.5, color=(1,0,0), edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, data['state_name'])
plt.xlim(1, max(data['positive']) + 10)
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

### Active Number of Cases based on States in India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,data['active'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, data['state_name'])  
plt.xlim(1,max(data['active'])+10) 
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

### Total Number of Deaths based on States in India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,data['death'], align='center', alpha=0.5,  
                 color=(0,0,1),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, data['state_name'])  
plt.xlim(1,max(data['death'])+10) 
plt.xlabel('Death Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```
